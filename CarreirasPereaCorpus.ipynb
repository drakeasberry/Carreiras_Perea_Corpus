```python
# import all needed modules
import PyPDF2
import re
import pandas as pd

# pdf file object
# you can find find the pdf file with complete code in below
pdfFileObj = open('DICCIONARIODEFRECUENCIASVOL1.pdf', 'rb')
# pdf reader object
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
# number of pages in pdf
#print(pdfReader.numPages)

startPageNum = 125
paperStartPageNum = startPageNum - 7
pagenum = startPageNum
words = []
freq = []
realPage = []

while pagenum < pdfReader.numPages: 
    # a page object
    pageObj = pdfReader.getPage(pagenum)
    #print(pagenum)
    # extracting text from page.
    # this will print the text you can also save that into String
    #print(pageObj.extractText())
    pageText=pageObj.extractText()
    #print(pageText)
    
    noDoubleSpace = re.sub(r' +',' ',pageText)
    noOddChar = re.sub(r'[-.]', ' ', noDoubleSpace)
    newTextbyLine = re.sub(r'\s','\n',noOddChar)
    #print(newTextbyLine)
    noColon = re.sub(r'[:]','',newTextbyLine)
    #print(new2)
    noLetterLabels = re.sub(r'[A-Z]','',noColon)
    #print(noLetterLabels)
    #noLetterLabels
    
    #Testing may need to delete
    removeempties = ''.join(noLetterLabels).split('\n')
    #print(removeempties)
    removeempties = ' '.join(removeempties).split()
    #print(removeempties)
    x = len(removeempties)
    popX = x - 1
    removeempties = removeempties[0:popX]
    #End Testing
    
    ####lines = noLetterLabels.split('\n')
    ####x = len(lines)
    ####popX = x -2
    #print(lines[0:popX])
    ####noPage = lines[0:popX]
    #print(noPage)
    ####removeempties=' '.join(noPage).split()
    #print(removeempties)
    #print('\nOriginal List Length',len(removeempties))
    #print(type(removeempties))
    
    i = 0
    while i < len(removeempties):
        #print('The length of the list is now',len(removeempties),'\n')
        if i % 2 == 0:
            pass
            #print(i,': is even and should be text -- ', removeempties[i])
        else:
            #print(i,': is odd and should be a number --', removeempties[i])
            if removeempties[i].isdigit() == True:
                pass
                #print('\t\t      Value equals: ',removeempties[i], ' is a number\n')
            else:
                if removeempties[i] == 'l':
                    #print('Im in the if code block')
                    #print('Value of current item is: ',removeempties[i], i)
                    removeempties[i] = '1'
                    #print(i)
                    #print('Replaced value of current item is: ',removeempties[i])
                    i = i-2
                else:
                    #print(i,': somethings is not right', removeempties[i], ' is not a number')
                    mini = i-1
                    maxi = i+1
                    #print('Before join:\n', '\tPrevious value = ',removeempties[mini], '\n\tCurrent value = ', removeempties[i], '\n\tNext value = ',removeempties[maxi],'\n')
                    removeempties[mini:maxi] = [''.join(removeempties[mini:maxi])]
                    #print('After join:\n', '\tPrevious value = ',removeempties[mini], '\n\tCurrent value = ', removeempties[i], '\n\tNext value = ',removeempties[maxi],'\n')
                    i = i-2
        i += 1
        #print(i)
    #print(removeempties)
    #print('Updated List Length',len(removeempties),'\n')

    k = 0

    for item in removeempties:
        if k % 2 == 0:
            words.append(removeempties[k])
            realPage.append(paperStartPageNum)
        else:
            freq.append(removeempties[k])
            #realPage.append(pagenum)  
        k += 1
    pagenum += 1
    paperStartPageNum +=1
    #print(words)
    #print(freq)
    #print('List length from page ',startPageNum,' to ', pagenum)
    #print('\nthe length of words is: ', len(words))
    #print('the length of frequency is: ', len(freq), '\n')
print(len(realPage))
print(len(words))
print(len(freq))

# Create a zipped list of tuples from above lists
zippedList =  list(zip(words, freq, realPage))
#print(zippedList)

dfObj = pd.DataFrame(zippedList, columns =['word','frequency','page']) 
#dfObj
```
